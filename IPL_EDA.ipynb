```python
def is_out_or_not(row):
    if (pd.notna(row['player_dismissed']) or pd.notna(row['dismissal_kind']) or pd.notna(row['fielder'])):
        return 1 
    else :
        return 0 

df['is_out_or_not']=df.apply(is_out_or_not,axis=1)
```

```python
df1
```

```python
df2=matches[['id','Season']]
```

```python
df2
```

```python
df3=pd.merge(df1,df2,on='id')
df3
```

```python
df4=df3.groupby('Season')['total_runs'].mean()
df4=pd.DataFrame(df4)
df4=df4.reset_index()
df4.columns=['Season','Average Runs']
df4
```

```python
ax=df4.plot(x='Season',kind='line',figsize=(12,6))
plt.xticks(ticks=range(len(df4['Season'])),labels=df4['Season'],rotation=45)
for i, val in enumerate(df4['Average Runs']):
    ax.text(i, val + 2, f'{val:.1f}', ha='center', fontsize=9, color='black')


plt.title('Average Runs in Every season of IPL')
```
